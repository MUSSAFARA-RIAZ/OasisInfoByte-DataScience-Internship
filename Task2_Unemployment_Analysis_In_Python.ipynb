```python
import plotly.express as px
unemployment = df_new.groupby(['Region', 'Region.1'])[' Estimated Unemployment Rate (%)'].mean().reset_index()

figure = px.sunburst(unemployment,
                     path=['Region', 'Region.1'],
                     values=' Estimated Unemployment Rate (%)',
                     color_continuous_scale='Electric',
                     title='Unemployment rate in each region and city',
                     height=650,
                     template='ggplot2')

figure.show()
```
